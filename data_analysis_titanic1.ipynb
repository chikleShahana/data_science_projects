```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"D:\shan_data science\Titanic-Dataset.csv", encoding='unicode_escape')
```

```python
dataset.head()
```

```python
am=np.mean(dataset['Age'])
```

```python
dataset['Age'].mean()
```

```python
amd=np.median(dataset['Age'])
```

```python
amod=dataset['Age'].mode()[0]
```

```python
sns.histplot(x='Age', data=dataset , bins=[i for i in range(0,81,10)])
plt.show
```

```python
sns.histplot(x='Age', data=dataset , bins=[i for i in range(0,100,10)])
plt.plot([am for i in range(0,300,10)],[i for i in range(0,300,10)],c='red')
plt.plot([amd for i in range(0,80,10)],[i for i in range(0,80,10)])
plt.plot([amod for i in range(0,300,10)],[i for i in range(0,300,10)] , c='green')
plt.show
                                    
```
